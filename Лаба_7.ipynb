```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x = np.arange(0,100,0.1)
y = 3 * np.power(x, 5) * np.sin(3* x + 2)
plt.figure(figsize=(10,7))
plt.plot(x,y)
```

```python
x = np.array([i for i in range(0, 11)])
y = np.sin(x)
plt.figure(figsize=(10,8))
plt.bar(x, y)
```

```python
x1 = np.random.normal(loc = 1.0, scale = 0.5, size=1000)
y1 = np.random.normal(loc = 1.0, scale = 0.5, size=1000)

x2 = np.random.normal(loc = 4.0, scale = 0.8, size=1000)
y2 = np.random.normal(loc = 4.0, scale = 0.8, size=1000) + np.sin(3*x2)

plt.figure(figsize=(12,8))
plt.scatter(x1, y1)
plt.scatter(x2, y2)
```

```python
t=np.arange(0, 360, 0.2)

x=4*np.cos(4*t) + 6*np.cos(5*t)
y=7*np.sin(1.5*t) + 5*np.sin(7*t)


plt.figure(figsize=(12,8))
plt.plot(x,y)
```

```python
import csv

x = []
y = []

with open("p4_data_24.csv", "r", newline='') as file:
    reader = csv.DictReader(file)
    for row in reader:
        try:
            price = float(row['Price'].replace('$', ''))
            x.append(row['Name'])
            y.append(price)
        except ValueError as e:
            print(f"Error converting price in row {row}: {e}")
        except KeyError as e:
            print(f"Error: Missing key {e} in row {row}")


print("Names:", x)
print("Prices:", y)
```
