```python
my_list = [1, 3, 5, "abc", 7, 9, 11, 2, 4, 13, 15, 17, 19.5]
```

```python
def count_odd_triples(data):

    count = 0
    n = len(data)

    for i in range(n - 2):

        if (isinstance(data[i], int) and data[i] % 2 != 0 and
            isinstance(data[i+1], int) and data[i+1] % 2 != 0 and
            isinstance(data[i+2], int) and data[i+2] % 2 != 0):
            count += 1

    return count


triple_count = count_odd_triples(my_list)

print(f"Количество троек подряд идущих нечетных чисел: {triple_count}")
```
